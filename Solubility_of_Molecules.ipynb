```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

```python
#data separation
y = df['logS']
y
```

```python
x = df.drop('logS' , axis=1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test, y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
#linear regression
#training the model

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#prediction

y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train_pred, y_lr_test_pred)
```

```python
y_lr_test_pred
```

```python
#model performance

from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2= r2_score(y_test,y_lr_test_pred)
```

```python
print("test MSE =", lr_test_mse)
print("test r2 =" ,lr_test_r2)
print("train MSE =", lr_train_mse)
print("train r2 =" ,lr_train_r2)
```

```python
lr_results= pd.DataFrame(("LinearRegression",lr_test_mse,lr_test_r2,lr_train_mse,lr_train_r2)).transpose()
lr_results.columns = ['Method','Testing MSE','Testing r2','Trainig MSE', 'Training r2']
```

```python
lr_results
```

```python
#RandomForest

#training the model

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train,y_train)
```

```python
# make prediction


y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2= r2_score(y_test,y_rf_test_pred)
```

```python
rf_results= pd.DataFrame(("RandomForest",rf_test_mse,rf_test_r2,rf_train_mse,rf_train_r2)).transpose()
rf_results.columns = ['Method','Testing MSE','Testing r2','Trainig MSE', 'Training r2']
```

```python
rf_results
```

```python
#model comparison

df_models = pd.concat((lr_results,rf_results), axis=0).reset_index(drop=True)
df_models
```
